```python
import os
import io

with io.open(os.path.join("../data/", "language-never-random.txt"), encoding='utf8') as fin:
    text = fin.read()
```

```python
tokens = text.split()
# lower case
tokens = [token.lower() for token in tokens]
```

```python
tokens
```

```python
kni = KneserNeyInterpolated(order=4, discount=0.75)
kni.train(tokens=tokens)

generated_tokens = kni.generate(context=["adam", "kilgarriff"], max_length=20)
print(' '.join(generated_tokens))
```

```python
import pickle

with open(os.path.join("../model/", "kneser-ney-interpolated-4gram-adam-kilgariff-from-scratch.pkl"), "wb") as fout:
    pickle.dump(kni, fout)
```

```python
with open(os.path.join("../model/", "kneser-ney-interpolated-4gram-adam-kilgariff-from-scratch.pkl"), "rb") as fin:
    kni = pickle.load(fin)

generated_tokens = kni.generate(context=["adam", "kilgarriff"], max_length=20)
```

```python
print(' '.join(generated_tokens))
```
