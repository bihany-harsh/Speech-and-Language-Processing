```python
from bs4 import BeautifulSoup
import requests

url = "https://en.wikipedia.org/wiki/The_Lord_of_the_Rings"
request = requests.get(url)

soup = BeautifulSoup(request.text, "html.parser")

# get the text from the first 7 <p> tags
paragraphs = soup.find_all("p")[:7]

text = ""

for paragraph in paragraphs:
    text += paragraph.text

import re
text = re.sub(r'\[[^]]*\]', '', text)
text = re.sub('\n', '', text)
text = re.sub('J\. R\. R\.', 'JRR', text)
text = re.sub('[,;]', ' ', text)
text = re.sub('\.', ' .', text)

corpus = [text]
```

```python
word2vec = Word2Vec(corpus)
word2vec.preprocess_corpus()
word2vec.build_model()
history = word2vec.train()
```

```python
print(word2vec.get_similar_words("rings"))
print(word2vec.get_similar_words("lord"))
print(word2vec.get_similar_words("frodo"))
```
